## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.

# Drop useless ID-like columns
df_eda = df.drop(columns=['ID', 'Customer_ID', 'Name', 'SSN'], axis=1)

# Handle missing values
# - For numerical columns → fill with median
for col in df_eda.select_dtypes(include=['int64', 'float64']).columns:
    df_eda[col] = df_eda[col].fillna(df_eda[col].median())

# - For categorical columns → fill with mode
for col in df_eda.select_dtypes(include=['object']).columns:
    df_eda[col] = df_eda[col].fillna(df_eda[col].mode()[0])

# Drop 'Type_of_Loan' (too messy for visualization in first step)
df_eda = df_eda.drop(columns=['Type_of_Loan'], axis=1)

print("EDA Dataset Shape:", df_eda.shape)
df_eda.head()
````

````python
df_clean = df.copy()

# Example cleaning steps
df_clean.drop_duplicates(inplace=True)
df_clean.dropna(inplace=True)

# Now check the cleaned data
df_clean.info()
df_clean.describe()
````

### What all manipulations have you done and insights you found?

Manipulations Done

1. Dropped ID-like columns


*   Columns: ID, Customer_ID, Name, SSN
*   Reason: They are unique identifiers, do not provide predictive or analytical value, and would only add noise.


2. Handled Missing Values


*   Numerical Columns → filled with median (robust against outliers, preserves distribution).
*   Categorical Columns → filled with mode (most frequent category, keeps categories meaningful).

3.  Dropped Type_of_Loan


*   It had 6,261 unique values (too high-cardinality, messy text format).
*   For initial analysis, this column was dropped because it would clutter plots and be hard to encode meaningfully.
